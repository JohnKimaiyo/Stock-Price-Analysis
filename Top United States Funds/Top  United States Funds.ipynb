```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
stocks = ['VSMPX', 'FXAIX', 'VFIAX', 'AGTHX', 'SWPPX']
years = [2023]
```

```python
stock_data = pd.DataFrame()

for stock in stocks:
    for year in years:
        data = yf.download(stock, start=f'{year}-01-01', end=f'{year}-12-31')['Close']
        data.rename(f"{stock}_{year}", inplace=True)
        stock_data = pd.concat([stock_data, data], axis=1)
```

```python
stock_data.plot(figsize=(12, 8))
plt.title('Top 5 United States - Funds Prices Comparison')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.show()
```
